```python
import pandas as pd
import numpy as np
```

#### Import Data

```python
df = pd.read_csv('../raw_data/FM_data_21.csv', low_memory = False)
```

```python
df.head()
```

```python
df.rename(str.strip, axis='columns', inplace = True)
```

```python
df["Preferred Foot"] = df["Preferred Foot"].str.strip()
```

```python
df['Preferred Foot'] = df['Preferred Foot'].replace("Right Only", "Right")
df['Preferred Foot'] = df['Preferred Foot'].replace("Left Only", "Left")
```

```python
# Drop rows where "Preferred Foot" is blank
df = df[df["Preferred Foot"] != '']

# Reset index if you want
df = df.reset_index(drop=True)

# Display the updated DataFrame
df.head()
```

```python
df["Preferred Foot"].unique()
```

```python
df["Left Foot"] = df["Left Foot"].str.strip()
df["Right Foot"] = df["Right Foot"].str.strip()
```

```python
df["Preferred Foot"] = np.where((df["Left Foot"] == "Very Strong") & (df["Preferred Foot"] == "Either"), "Either_left", df["Preferred Foot"])
df["Preferred Foot"] = np.where((df["Right Foot"] == "Very Strong") & (df["Preferred Foot"] == "Either"), "Either_right", df["Preferred Foot"])
```

```python
df["Preferred Foot"].unique()
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Replace "Preferred Foot" with your actual column name
# If you haven't already removed leading and trailing whitespaces, you can use df["Preferred Foot"] = df["Preferred Foot"].str.strip() before encoding

# Perform one-hot encoding and convert to integers (0 or 1)
df_encoded = pd.get_dummies(df["Preferred Foot"], prefix="Preferred_Foot").astype(int)

# Concatenate the new one-hot encoded columns to the original DataFrame
df = pd.concat([df, df_encoded], axis=1)

# Drop the original "Preferred Foot" column if needed
df = df.drop("Preferred Foot", axis=1)

```
