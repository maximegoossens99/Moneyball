```python
import pandas as pd
```

```python
df = pd.read_csv("../raw_data/clean_data.csv")
```

```python
df.columns

# Step 1: Drop rows where the club is equal to 'Other'
df = df[df['club'] != 'Other']

def filter_position(df, club, position, max_count):
    filtered_df = df[(df['club'] == club) & (df[position] == 1)]
    return filtered_df.head(max_count)

def filter_and_append(avg_clubs, df, club, position, max_count):
    filtered_df = filter_position(df, club, position, max_count)
    return pd.concat([avg_clubs, filtered_df])

avg_clubs = pd.DataFrame()

for club in df['club'].unique():
    avg_clubs = filter_and_append(avg_clubs, df, club, 'goalkeeper', 2)
    avg_clubs = filter_and_append(avg_clubs, df, club, 'centerback', 3)
    avg_clubs = filter_and_append(avg_clubs, df, club, 'fullback', 4)
    avg_clubs = filter_and_append(avg_clubs, df, club, 'winger', 4)
    avg_clubs = filter_and_append(avg_clubs, df, club, 'midfielder', 5)
    avg_clubs = filter_and_append(avg_clubs, df, club, 'striker', 2)
```

```python
df = avg_clubs.groupby(by='club').mean()

df.sort_values(by='club_rating', ascending=False, inplace=True)
```

```python
def create_grouped_features(df):
    # Offensive Skills
    df['shooting'] = df[['fin', 'lon', 'fre', 'pen']].mean(axis=1)
    df['dribbling_control'] = df[['dri', 'fir', 'fla', 'tec']].mean(axis=1)
    df['passing_vision'] = df[['pas', 'vis', 'l th', 'cro', 'cor', 'otb']].mean(axis=1)
    
    # Defensive Skills
    df['tackling_interception'] = df[['tck', 'mar', 'pos', 'ant']].mean(axis=1)
    df['aerial_defense'] = df[['hea', 'jum', 'aer']].mean(axis=1)
    
    # Physical Attributes
    df['speed_agility'] = df[['acc', 'pac', 'agi']].mean(axis=1)
    df['strength_stamina'] = df[['str', 'sta', 'bal']].mean(axis=1)
    
    # Mental and Tactical Attributes
    df['decision_making'] = df[['dec', 'cmp']].mean(axis=1)
    df['work_ethic_effort'] = df[['wor', 'det', 'bra']].mean(axis=1)
    
    # Leadership and Teamwork
    df['leadership'] = df[['ldr', 'com']].mean(axis=1)
    df['teamwork'] = df['tea']
    
    # Goalkeeping Abilities
    df['goalkeeping_abilities'] = df[['han', 'ref', 'kic', 'thr', '1v1', 'ecc', 'cmd']].mean(axis=1)
    
    # Behavioral Attributes
    df['behavioral_attributes'] = df[['agg', 'pun']].mean(axis=1)
    
    # Player Traits
    df['player_traits'] = df['tro']
    
    # Create a new DataFrame with the grouped features
    dfgf = df[['shooting', 'dribbling_control', 'passing_vision', 'tackling_interception', 
               'aerial_defense', 'speed_agility', 'strength_stamina', 'decision_making', 
               'work_ethic_effort', 'leadership', 'teamwork', 'goalkeeping_abilities', 
               'behavioral_attributes', 'player_traits', 'club_rating']]
    
    return dfgf
```

```python
df = create_grouped_features(df)
```

```python
counter_attacking_style = {
    'relevant_features': ['speed_agility', 'passing_vision', 'tackling_interception', 'shooting'],
    'weights': [0.40, 0.13, 0.12, 0.35]
}

# High-Pressing Havoc
high_pressing_style = {
    'relevant_features': ['speed_agility', 'tackling_interception', 'work_ethic_effort', 'decision_making'],
    'weights': [0.30, 0.25, 0.20, 0.25]
}

# Defensive Fortress
defensive_style = {
    'relevant_features': ['aerial_defense', 'strength_stamina', 'work_ethic_effort', 'tackling_interception'],
    'weights': [0.40, 0.18, 0.17, 0.25]
}

# Wing Dominance
wing_dominance_style = {
    'relevant_features': ['speed_agility', 'strength_stamina', 'decision_making', 'shooting'],
    'weights': [0.50, 0.10, 0.20, 0.20]
}

# Possession with Purpose
possession_with_purpose_style = {
    'relevant_features': ['teamwork', 'dribbling_control', 'passing_vision', 'decision_making'],
    'weights': [0.10, 0.35, 0.30, 0.25]
}

# Youthful Energy and High Intensity
youthful_energy_style = {
    'relevant_features': ['dribbling_control', 'speed_agility', 'strength_stamina', 'tackling_interception'],
    'weights': [0.35, 0.40, 0.10, 0.15]
}

# Midfield Maestros
midfield_maestros_style = {
    'relevant_features': ['decision_making', 'leadership', 'player_traits', 'behavioral_attributes'],
    'weights': [0.35, 0.20, 0.35, 0.10]
}
```

```python
styles = {
    'counter_attack': counter_attacking_style, 
    'high_pressing': high_pressing_style, 
    'defensive': defensive_style, 
    'midfield_maestros': midfield_maestros_style,
    'wing_dominance': wing_dominance_style, 
    'possession_with_purpose': possession_with_purpose_style,
    'youthful_energy': youthful_energy_style}
```

```python
def calculate_team_score(df, playing_style):
    relevant_features = playing_style['relevant_features']
    weights = playing_style['weights']

    # Filter DataFrame to include only relevant features
    relevant_df = df[relevant_features]

    # Normalize the relevant features to have values between 0 and 1
    normalized_df = (relevant_df - relevant_df.min()) / (relevant_df.max() - relevant_df.min())

    # Calculate the weighted sum for each team
    team_scores = (normalized_df * weights).sum(axis=1)
    
    # Making a DataFrame
    team_scores = pd.DataFrame(team_scores).sort_values(by=0, axis=0, ascending=False)

    # Get the team with the highest score
    best_team = team_scores.idxmax()

    return best_team, team_scores

# Create a DataFrame to store style scores for each team
all_style_scores = pd.DataFrame(index=df.index)

# Calculate the best team for each style and store the scores in the DataFrame
for style_name, style in styles.items():
    best_team, team_scores = calculate_team_score(df, style)
    
    # Renaming column
    team_scores = team_scores.rename(columns={0: style_name})
    
    # Merge the style scores with the overall DataFrame
    all_style_scores = pd.merge(all_style_scores, team_scores, left_index=True, right_index=True, how='outer')

    print(f"Best team for {style_name}: {best_team}")

# Calculate the average score for each team across all styles
all_style_scores['average_score'] = all_style_scores.mean(axis=1)

# Sort by the best average score
all_style_scores = all_style_scores.sort_values(by='average_score', ascending=False)


all_style_scores
```

```python
### positions dataframe

def custom_scaler(dfgf_no_name):
    total_score = dfgf_no_name.sum(axis=1)
    scaled_dfgf_no_name = dfgf_no_name.div(total_score, axis=0)
    return scaled_dfgf_no_name
```

```python
df = custom_scaler(df)
```

```python
# Create a DataFrame to store style scores for each team
all_style_scores = pd.DataFrame(index=df.index)

# Calculate the best team for each style and store the scores in the DataFrame
for style_name, style in styles.items():
    best_team, team_scores = calculate_team_score(df, style)
    
    # Renaming column
    team_scores = team_scores.rename(columns={0: style_name})
    
    # Merge the style scores with the overall DataFrame
    all_style_scores = pd.merge(all_style_scores, team_scores, left_index=True, right_index=True, how='outer')

    print(f"Best team for {style_name}: {best_team}")

# Calculate the average score for each team across all styles
all_style_scores['average_score'] = all_style_scores.mean(axis=1)

# Sort by the best average score
all_style_scores = all_style_scores.sort_values(by='average_score', ascending=False)


all_style_scores
```
